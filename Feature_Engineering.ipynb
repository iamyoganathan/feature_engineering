Task 1 — Categorical Encoding

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Original Data
data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'City': ['Mumbai', 'Delhi', 'Chennai', 'Mumbai', 'Delhi'],
    'Size': ['Small', 'Large', 'Medium', 'Large', 'Small'],
    'Age': [25, 45, 32, 60, 28],
    'Fare': [15, 300, 85, 450, 20]
}

df = pd.DataFrame(data)

# -------------------------------
# 1. Gender → Label Encoding
# -------------------------------
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
# Male=1, Female=0 (automatically assigned alphabetically)

# -------------------------------
# 2. City → One-Hot Encoding
# drop_first=True avoids dummy variable trap
# -------------------------------
df = pd.get_dummies(df, columns=['City'], drop_first=True)

# -------------------------------
# 3. Size → Ordered Label Encoding
# Small=0, Medium=1, Large=2
# -------------------------------
size_mapping = {'Small': 0, 'Medium': 1, 'Large': 2}
df['Size'] = df['Size'].map(size_mapping)

print("Encoded Data:\n", df)
```

Task 2 — Feature Scaling

```python
from sklearn.preprocessing import RobustScaler

# -------------------------------
# Using RobustScaler because:
# Fare has an outlier (450).
# RobustScaler uses median and IQR,
# so it is not affected by outliers.
# -------------------------------

scaler = RobustScaler()

df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])

print("\nScaled Data:\n", df)
```
